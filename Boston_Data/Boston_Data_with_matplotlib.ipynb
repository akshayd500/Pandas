```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Boston.csv")
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
colors = ["red" , "blue" , "green"]
plt.scatter(x , y , 500 , c = "red" , marker = "$-$" , alpha = .50 )
```

```python
plt.scatter(x,y,500,marker="+")
```

```python
data_crime = data[data.crim < 10]
x = data_crime.crim
y = data_crime.medv
plt.scatter(x,y,alpha=0.3)
```

```python
age_limit = 50
age_under = data [ ( data.age < age_limit ) & ( data.crim <10 ) ]
age_above = data [ ( data.age > age_limit ) & ( data.crim <10 ) ]

x_under = age_under.crim
y_under = age_under.medv

plt.scatter(x_under,y_under,c="red")

x_above = age_above.crim
y_above = age_above.medv

plt.scatter(x_above,y_above,alpha=0.2)
```
